```python
# Task 1 - Iris 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("iris_sample.csv")   
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
df.head()
```

```python
df.info()
df.describe()
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x='sepal_length', y='petal_length', hue='species', s=80)
plt.title('Sepal Length vs Petal Length')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['sepal_length'].dropna(), bins=8, kde=True)
plt.title('Distribution of Sepal Length')
plt.xlabel('sepal_length')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df[['sepal_length','sepal_width','petal_length','petal_width']])
plt.title('Boxplot of Numeric Features')
plt.show()
```
